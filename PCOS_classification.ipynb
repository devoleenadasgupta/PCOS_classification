```python
import openpyxl #helps read excel files
import pandas as pd 
data = pd.read_excel('PCOS_data_without_infertility.xlsx', sheet_name=1)

data.head(10) 
```

```python
data_inf = pd.read_csv('PCOS_infertility.csv')
data_inf.head(10)
```

```python
data['Patient File No.'] = data['Patient File No.'].apply(lambda x:  x+10000).astype('int64')
```

```python
data.head()
```

```python
#merging data from second table into first table:
data = pd.merge(data,data_inf[['Patient File No.','  I   beta-HCG(mIU/mL)','II    beta-HCG(mIU/mL)']], on='Patient File No.', how='left')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.describe().T
```

```python
del data['Unnamed: 44']#since 539 null values are there, we are deleting the column
data['Marraige Status (Yrs)'].fillna(0,inplace = True)#since we have only one null value, we're replacing it with 0
data['Fast food (Y/N)'].fillna(0,inplace = True)#since there is only one null value, we're replacing it with 0
data
```

```python
#plotting bar plots for various columns:
import matplotlib.pyplot as plt 
import seaborn as sns 
for i in [' Age (yrs)', 'Weight (Kg)',
       'Height(Cm) ', 'Hb(g/dl)', 'Cycle(R/I)', 'Cycle length(days)', 'No. of aborptions',
        'Hip(inch)', 'Waist(inch)', 
       'PRG(ng/mL)', 'RBS(mg/dl)', 'BP _Systolic (mmHg)', 'Follicle No. (L)', 'Follicle No. (R)',
       'Avg. F size (L) (mm)', 'Avg. F size (R) (mm)', 'Endometrium (mm)']:
  sns.set(rc = {'figure.figsize':(15,15)})
  data[data['PCOS (Y/N)'] == 1][i].value_counts().plot.bar()
  plt.title(i)
  plt.show()
```

```python
#Plotting correlation plots:
sns.heatmap(data.corr('pearson'), annot =True, annot_kws={'size': 5} )
sns.set(rc = {'figure.figsize':(70,70)})
plt.show()
```

```python
data = data[['PCOS (Y/N)','Follicle No. (R)','Follicle No. (L)','Weight gain(Y/N)','hair growth(Y/N)','Skin darkening (Y/N)','Cycle(R/I)','Fast food (Y/N)','Cycle length(days)',' Age (yrs)']]
sns.heatmap(data.corr('pearson'), annot =True, annot_kws={'size': 50} )
sns.set(rc = {'figure.figsize':(60,60)})
plt.show()
```

```python
data['PCOS (Y/N)'].value_counts()
```

```python
X=data.drop(["PCOS (Y/N)"],axis = 1)
y=data["PCOS (Y/N)"]
```

```python
#!pip install imbalanced-learn
```

```python
#oversampling:
from imblearn.over_sampling import RandomOverSampler 
oversample = RandomOverSampler(sampling_strategy=0.7)
X, y = oversample.fit_resample(X, y)
y.value_counts()
```

```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import math
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
sscaler = MinMaxScaler() #scaling the dataset. This will make it easy for the model to train
cols = X.columns
x_scaled = sscaler.fit_transform(X)
X_scaled = pd.DataFrame(x_scaled, columns = cols)
X_scaled
```

```python
X_scaled.describe().T
```

```python
X_scaled_values = X_scaled.to_numpy() #convert DataFrame to numpy array
X_scaled_values
```

```python
#test train split:
X_train,X_test, y_train, y_test = train_test_split(X_scaled , y, test_size=0.2) 
```

```python
#Applying decision tree classifier
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()

# Train the classifier
clf.fit(X_train, y_train)

# Predict on the test set
y_pred = clf.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
classi_report1 = classification_report(y_test, y_pred)
print(classi_report1)
```

```python
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, cmap="Blues", fmt="d")
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
#Applying random forest classifier:
rfc = RandomForestClassifier(n_jobs=-1,n_estimators=150,max_features='sqrt',min_samples_leaf=10) #creates a Random forest model
rfc.fit(X_train, y_train) 
pred_rfc = rfc.predict(X_test) 
accuracy = accuracy_score(y_test, pred_rfc)
print(accuracy)
```

```python
classi_report2 = classification_report(y_test, pred_rfc)
print(classi_report2)
```

```python
cm = confusion_matrix(y_test, pred_rfc)
print("Confusion Matrix:")
print(cm)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, cmap="Blues", fmt="d")
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```
